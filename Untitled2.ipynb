```python

class Patient:
    def __init__(self, patient_id, name, age, medical_history):
        self.patient_id = patient_id
        self.name = name
        self.age = age
        self.medical_history = medical_history

class Doctor:
    def __init__(self, doctor_id, name, specialization):
        self.doctor_id = doctor_id
        self.name = name
        self.specialization = specialization

    def get_doctor_id(self):
        return self.doctor_id

    def get_name(self):
        return self.name

    def get_specialization(self):
        return self.specialization

class Appointment:
    def __init__(self, appointment_id, patient, doctor, date, time):
        self.appointment_id = appointment_id
        self.patient = patient
        self.doctor = doctor
        self.date = date
        self.time = time

class HospitalManagementSystem:
    def __init__(self):
        self.medical_charges = 0  # Variable to hold medical charges
        self.doctors = [
            Doctor("D1", "Dr. Sampath", "Cardiologist"),
            Doctor("D2", "Dr. Arjun", "Pediatrician"),
            Doctor("D3", "Dr. Anjali", "Orthopedic Surgeon"),
            Doctor("D4", "Dr. Rahul", "Dermatologist"),
            Doctor("D5", "Dr. Sakshi", "Psychiatrist")
        ]

        self.patient_id = None
        self.patient_name = None
        self.patient_age = None
        self.doctor_id = None
        self.doctor_name = None
        self.appointment_date = None
        self.appointment_time = None

    def add_patient(self):
        while True:
            try:
                self.patient_id = int(input("Enter patient ID: "))
                break
            except ValueError:
                print("Invalid input. Please enter a valid integer for patient ID.")

        while True:
            self.patient_name = input("Enter patient name: ")
            if self.patient_name.isalpha():
                break
            else:
                print("Invalid input. Please enter a valid patient name containing only letters.")

        while True:
            try:
                self.patient_age = int(input("Enter patient age: "))
                if self.patient_age >= 0:
                    break
                else:
                    print("Invalid input. Age cannot be negative. Please enter a valid age:")
            except ValueError:
                print("Invalid input. Please enter a valid integer for patient age.")

        print("Enter patient's medical history: (mention if any or enter nil)")
        input()  # Consume medical history
        print("Patient added successfully.")

    def add_doctor(self):
        print("Select a doctor:")
        for i, doctor in enumerate(self.doctors):
            print(f"{i + 1}. {doctor.get_name()} ({doctor.get_specialization()})")

        while True:
            try:
                choice = int(input("Enter your choice: "))
                if 1 <= choice <= len(self.doctors):
                    break
                else:
                    print("Invalid choice. Please enter a valid choice from the list.")
            except ValueError:
                print("Invalid input. Please enter a valid integer choice.")

        doctor = self.doctors[choice - 1]
        self.doctor_id = doctor.get_doctor_id()
        self.doctor_name = doctor.get_name()
        print("Doctor selected successfully.")

    def schedule_appointment(self):
        while True:
            self.appointment_id = input("Enter appointment ID: ")
            if self.appointment_id:
                break
            else:
                print("Appointment ID cannot be empty. Please enter a valid appointment ID:")

        self.appointment_date = input("Enter appointment date (YYYY-MM-DD):")
        self.appointment_time = input("Enter appointment time (HH:MM):")

        print("Appointment scheduled successfully.")

    def add_medical_charge(self):
        while True:
            try:
                self.medical_charges = int(input("Enter the medical charges: "))
                break
            except ValueError:
                print("Invalid input. Please enter a valid integer for medical charges.")

    def generate_bill(self):
        print("-----------Generated Bill-----------")
        print("Patient ID:", self.patient_id)
        print("Patient Name:", self.patient_name)
        print("Patient Age:", self.patient_age)
        print("Doctor ID:", self.doctor_id)
        print("Doctor Name:", self.doctor_name)
        print("Appointment Date:", self.appointment_date)
        print("Appointment Time:", self.appointment_time)
        print("Medical Charges: $", self.medical_charges)

def main():
    hospital_system = HospitalManagementSystem()

    while True:
        print("\nHospital Management System")
        print("1. Add Patient")
        print("2. Add Doctor")
        print("3. Schedule Appointment")
        print("4. Add Medical Charge")
        print("5. Generate Bill")
        print("6. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            hospital_system.add_patient()
        elif choice == "2":
            hospital_system.add_doctor()
        elif choice == "3":
            hospital_system.schedule_appointment()
        elif choice == "4":
            hospital_system.add_medical_charge()
        elif choice == "5":
            hospital_system.generate_bill()
        elif choice == "6":
            print("Exiting...")
            return
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
